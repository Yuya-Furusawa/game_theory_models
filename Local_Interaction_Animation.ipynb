```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
plt.rcParams["animation.html"] = "jshtml"
import random
from matplotlib.animation import FuncAnimation, ArtistAnimation
from localint import LocalInteraction
from IPython.display import Image
import io
import base64
from IPython.display import HTML
```

```python
%%capture

coordination_game = np.array([[11, 0],
                              [9, 8]])
node_colors_2 = ['b', 'y']
N = 100
circle = {}
G = nx.cycle_graph(n=N)
circle['G'] = G
circle['adj_matrix'] = nx.adjacency_matrix(G)
circle['pos'] = nx.circular_layout(G)

li_coor = LocalInteraction(coordination_game, circle['adj_matrix'])

init_actions = np.zeros(li_coor.N, dtype=int)
init_actions[[0, -1]] = 1
li_coor.set_init_actions(init_actions)

num_actions = li_coor.num_actions

fig, ax = plt.subplots(figsize=(14,9), facecolor='w')
plt.axis('off')


def get_fig(n, ax):
    ax
    nx.draw_networkx_edges(G, pos=circle['pos'], alpha=0.5, width=2, arrows=False)
    for i in range(num_actions):
        nodelist = np.where(li_coor.current_actions==i)[0].tolist()
        nx.draw_networkx_nodes(G=circle['G'], pos=circle['pos'], node_size=200, nodelist=nodelist, node_color=node_colors_2[i])
    li_coor.play()


anim = FuncAnimation(fig, get_fig, fargs=(ax,), interval=100, frames=50)
```
